This notebook aims to visualize the same histogram plot using different seaborn styles: `whitegrid`, `darkgrid`, `white`, `dark`, and `ticks`.

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)
```

Let's create a DataFrame consisting of two columns: Floats and Categories.
- **Floats:** Contains `1,000` values drawn from a standard normal distribution.
- **Categories:** Contains a list of `1,000` randomly generated items, either Group A or Group B.

```python
# Create the DataFrame
df = pd.DataFrame({
    'Floats': np.random.randn(1000),
    'Categories': np.random.choice(['Group A', 'Group B'], size=1000)
})

# Display the first 3 rows of the DataFrame
df.head(3) 
```

```python
# Create a figure with a specific size, layout, and resolution
fig = plt.figure(figsize=(12, 10), constrained_layout=True, dpi=300)

# Add a super title for the entire figure
fig.suptitle('Histogram Plot Using Different Seaborn Styles', fontsize=16)

# Set global Seaborn styling parameters
sns.set(palette='deep', 
        context='notebook', 
        font='serif', 
        font_scale=1.1)

# List of Seaborn styles to apply
styles = ['whitegrid', 'darkgrid', 'white', 'dark', 'ticks']

# Define a function to plot a histogram with a specific style
def plot_with_style(dataframe, column, the_style, place=321):
    with sns.axes_style(the_style): # Apply a specific Seaborn style
        ax = fig.add_subplot(place) # Add subplot to the figure at the specified position
        
        # Create a histogram with Seaborn
        sns.histplot(dataframe[column], 
                     bins=20, 
                     kde=False, 
                     color='skyblue', 
                     edgecolor='black')

        # Add title and labels to the subplot
        ax.set_title(f"Style: {the_style.capitalize()}", fontsize=12)
        ax.set_xlabel("Values")
        ax.set_ylabel("Frequency")

        # set custom x-axis ticks
        ax.set_xticks(ticks=np.arange(-5, 6, 1)) # Range: -5 to 5 with step 1
    

# Loop through styles and their corresponding subplot positions
for the_style, the_place in zip(styles, [321, 322, 323, 324, 325]):
    plot_with_style(df, 'Floats', the_style, the_place) # Call the function for each style

plt.show() # Display the entire figure
```
