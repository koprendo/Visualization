```python
# Create the DataFrame
df = pd.DataFrame({
    'Floats': np.random.randn(1000),
    'Categories': np.random.choice(['Group A', 'Group B'], size=1000)
})

# Display the first 3 rows of the DataFrame
df.head(3) 
```

```python
def plot_with_style(the_style, place=321):
    with sns.axes_style(the_style):
        ax = fig.add_subplot(place)
        sns.histplot(df.Floats, 
             bins=20, 
             kde=False, 
             color='skyblue', 
             edgecolor='black'
        )

        # add plot title and labels
        ax.set_title(f"Style: {the_style}", fontsize=12)
        ax.set_xlabel("Values")
        ax.set_ylabel("Frequency")

        # set custom x-axis ticks
        ax.set_xticks(ticks=np.arange(-5, 6, 1))
    
fig = plt.figure(figsize=(12,10), constrained_layout=True, dpi=300)

sns.set(palette='deep',
        context='notebook',
        font='serif', 
        font_scale=1.1)

styles = ['whitegrid', 'darkgrid', 'white', 'dark', 'ticks', 'ticks']

for the_style, the_place in zip(styles, [321, 322, 323, 324, 325]):
    plot_with_style(the_style, the_place)

plt.show()
```
