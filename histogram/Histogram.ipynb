This notebook aims to visualize basic histograms using synthetic datasets.

Let's create a DataFrame consisting of two columns: Floats and Categories.
- **Floats:** Contains `1,000` values drawn from a standard normal distribution.
- **Categories:** Contains a list of `1,000` randomly generated items, either Group A or Group B.

```python
# Create the DataFrame
df = pd.DataFrame({
    'Floats': np.random.randn(1000),
    'Categories': np.random.choice(['Group A', 'Group B'], size=1000)
})

# Display the first 3 rows of the DataFrame
df.head(3) 
```

```python
# set the style for the plot
sns.set(style='darkgrid')

# create a figure with specified size and resolution
plt.figure(figsize=(12, 4), dpi=300)

# plot a histogram using seaborn
sns.histplot(df.Floats, 
             bins=20, 
             kde=False, 
             color='skyblue', 
             edgecolor='black'
)

# add plot title and labels
plt.title('Histogram', fontsize= 14, pad=20)
plt.ylabel('Counts', fontsize= 12, labelpad=15)
plt.xlabel('Floats', fontsize= 12, labelpad=15)

# set custom x-axis ticks
plt.xticks(ticks=np.arange(-5, 6, 1))

# display the plot
plt.show()
```
